Numpy is able to import text data files.

```python
import numpy as np
data = np.loadtxt("data/sample_nohead.txt")
print("data : {}\n".format(type(data)), data)
```

```python
import io
signal_raw = io.StringIO(open('data/sample_signal_4k.txt').read().replace(',','.'))
signal = np.loadtxt(signal_raw, skiprows=5)
signal
```

```python
signal_raw.seek(0)
time, ecg, Z = np.loadtxt(signal_raw, skiprows=5, unpack=True)
print("time = ", time)
print("ecg  = ", ecg)
print("Z    = ", Z)
```

Scipy is able to import directly MATLAB files

```python
import scipy.io as sio
MAT_FILE = "./data/matlab_file.mat"
print("Contents of ", MAT_FILE)
for var in sio.whosmat(MAT_FILE):
    print(var)
data = sio.loadmat(MAT_FILE)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
Err = data['Err_tableau']
plt.imshow(Err, aspect=0.5)
plt.title("Deformation map $\epsilon_{rr}$", size=18);
plt.colorbar();
```
